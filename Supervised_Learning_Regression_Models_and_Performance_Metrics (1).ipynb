1)  What is Simple Linear Regression (SLR)? Explain its purpose?
- Simple Linear Regression (SLR) is a statistical method used to model the relationship between:

a) One independent variable (X)

b) One dependent variable (Y)
- It assumes that the relationship between X and Y is linear, meaning it can be represented by a straight line.
- Mathematical form :
$$
Y = \beta_0 + \beta_1 X + \varepsilon
$$

Where:

- $Y$ = Dependent variable (target/output)  
- $X$ = Independent variable (predictor/input)  
- $\beta_0$ = Intercept (value of $Y$ when $X=0$)  
- $\beta_1$ = Slope (change in $Y$ for one unit change in $X$)  
- $\varepsilon$ = Error term
- Purpose of Simple Linear Regression :

i) To Understand Relationship : SLR helps to determine whether a relationship exists between two variables and whether the relationship is positive, negative, or no relationship. e.g : Study hours and exam marks.

ii) To Predict Values : It is used to predict the value of Y for a given value of X.e.g: predict salary based on years of experience.

iii) To Measure Strength of Impact : The slope tells how much Y changes when
X increases by 1 unit and also  magnitude and direction of influence.

iv) To Minimize Error : SLR finds the best fitting line using the Least Squares Method, which minimizes:$$
\sum (Y - \hat{Y})^2
$$
This ensures the prediction line is as close as possible to actual data points.

2) What are the key assumptions of Simple Linear Regression?
- For Simple Linear Regression to give reliable results, the following assumptions must be satisfied:

i) Linearity : There must be a linear relationship between the independent variable X and dependent variabl Y. The data should follow a straight-line pattern.

ii) Independence of Errors : he error terms (ε) must be independent of each other i.e One observation should not influence another.

iii) Constant Variance : The variance of errors should be constant for all values of X : $$
\mathrm{Var}(\varepsilon) = \sigma^2
$$
The spread of residuals should be uniform.

iv) Normality of Errors : The error terms should be normally distributed:$$
\varepsilon \sim N(0, \sigma^2)
$$
Important for hypothesis testing and confidence intervals.

v) No Perfect Multicollinearity : Since SLR has only one independent variable, this assumption is automatically satisfied.

3) Write the mathematical equation for a simple linear regression model and
explain each term?
- The mathematical form of the Simple Linear Regression (SLR) model is:
$$
Y = \beta_0 + \beta_1 X + \varepsilon
$$

 Y — Dependent Variable : Also called the response variable or target variable.It is the variable we want to predict or explain.

X — Independent Variable : Also called the predictor variable or explanatory variable.It is the variable used to explain changes in Y.e.g: Advertising cost, study hours, years of experience.

 $\beta_0$ — Intercept : The value of Y when X=0 .It represents the starting point of the regression line.Interpretation: Baseline value of Y.

 $\beta_1$ — Slope (Regression Coefficient) : Represents the change in Y for a one-unit increase in X.Shows the direction and strength of the relationship.

If $\beta_1 > 0$ → Positive relationship  

If $\beta_1 < 0$ → Negative relationship  

If $\beta_1 = 0$ → No linear relationship
- Error Term : Represents the difference between the actual value and the predicted value.  Accounts for randomness and factors not included in the model. $$
\varepsilon = Y - \hat{Y}
$$

5) What is the method of least squares in linear regression?
- The Method of Least Squares is a statistical technique used to estimate the parameters of a linear regression model.
- It determines the best-fitting line by minimizing the sum of the squared differences between the actual values and the predicted values.
- Mathematical Idea :

 In Simple Linear Regression, the model is:
$$Y = \beta_0 + \beta_1 X + \varepsilon$$
The predicted value is:
$$\hat{Y} = \beta_0 + \beta_1 X$$
The error (residual) is:$$e_i = Y_i - \hat{Y}_i$$
The method minimizes the Sum of Squared Errors :$$S = \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2$$
- Errors are squared because:

To remove negative signs

To penalize larger errors more heavily

To ensure a unique minimum solution.
- Objective Function :The goal is to find $\beta_0$ and $\beta_1$ such that:
$$\min_{\beta_0,\beta_1} \sum_{i=1}^{n} (Y_i - \beta_0 - \beta_1 X_i)^2$$
- Estimated Coefficients : The least squares estimates are:$$\hat{\beta}_1 =
\frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{\sum (X_i - \bar{X})^2}$$
- Interpretation :

Minimizes total prediction error

Passes as close as possible to all data points

Provides the best linear unbiased estimates under assumptions.

6) What is Logistic Regression? How does it differ from Linear Regression?
- Logistic Regression is a supervised learning algorithm used for classification problems, especially when the dependent variable is binary (0 or 1).
- It predicts the probability that an observation belongs to a particular class.
- Mathematical Model :Instead of predicting Y directly, Logistic Regression predicts probability:$$P(Y=1) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X)}}$$
This function is called the Sigmoid (Logistic) Function.
- How Logistic Regression Differs from Linear Regression :
|Feature |Linear Regression|Logistic Regression|
|-------------------|---------------------|----------------------------------|
|Type of Problem|Regression|Classification|
|Output|Any real number|Probability (0 to 1)|
|Model Equation|$$Y = \beta_0 + \beta_1 X$$|$$P(Y=1) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X)}}$$|
|Error Minimization|Least Squares|Maximum Likelihood|
|Assumes Normal Errors|Yes|No|
|Decision Boundary|Straight line|S-shaped|

7) Name and briefly describe three common evaluation metrics for regression
models.
- i) Mean Absolute Error (MAE) :
 Formula: $$MAE = \frac{1}{n} \sum_{i=1}^{n} |Y_i - \hat{Y}_i|$$
- Description:

Measures the average absolute difference between actual and predicted values.

Errors are not squared.

Easy to interpret because it is in the same unit as the dependent variable.
- Interpretation: :

Smaller MAE → Better model performance.

MAE = 0 means perfect prediction.
- ii) Mean Squared Error (MSE) :Formula:$$MSE = (1/n) Σ(i=1 to n) (Yi − Ŷi)^2
$$
- Description:

Measures the average squared difference between actual and predicted values.

Penalizes larger errors more heavily due to squaring.

Useful when large errors are particularly undesirable.
- Interpretation:

Smaller MSE → Better model.

Sensitive to outliers.
- Root Mean Squared Error L :Formula:$$RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2}$$
- Description:

Square root of MSE.

Brings error back to the original unit of the dependent variable.

Most commonly used regression metric.
- Interpretation:

Lower RMSE → Better predictive accuracy.

Sensitive to large errors.

8) What is the purpose of the R-squared metric in regression analysis?
- R-squared measures how well a regression model explains the variability of the dependent variable.
- Mathematical Definition :
$$R^2 = 1 - \frac{SS_{res}}{SS_{tot}}$$
where ,
$$SS_{res} = \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2$$
$$SS_{tot} = \sum_{i=1}^{n} (Y_i - \bar{Y})^2$$
- Interpretation :

$$R^2 = 1 \rightarrow \text{Perfect fit}$$
$$R^2 = 0 \rightarrow \text{Model explains none of the variability}$$
$$R^2 = 0.85 \rightarrow \text{Model explains 85\% of the variation}$$
$$R^2 < 0 \rightarrow \text{Model performs worse than predicting the mean}$$
- Why It Is Important :
Measures goodness of fit

Helps compare different regression models

Easy to interpret as a percentage.
- Limitation :

Always increases when more features are added.

That is why Adjusted $R^2$ is preferred in multiple regression models.

9) Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept?








```python

import numpy as np
from sklearn.linear_model import LinearRegression

# Independent variable (X)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # must be 2D

# Dependent variable (Y)
Y = np.array([2, 4, 5, 4, 5])

# Creating Linear Regression model
model = LinearRegression()

# Fit the model
model.fit(X, Y)

# Get the slope (coefficient) and intercept
slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope (Coefficient): {slope}")
print(f"Intercept: {intercept}")


#output :
#Slope (Coefficient): 0.6
#Intercept: 2.2

```
